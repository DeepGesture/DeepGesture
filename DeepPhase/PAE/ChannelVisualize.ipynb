```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_channels = 5

phases = []
frequencies = []
amplitudes = []
offsets = []
```

```python
with open("./Training/Parameters_10.txt") as file:
    for line in file.readlines():
        params = line.split(" ")
        for i in range(n_channels):
            phase = params[0 * n_channels + i]
            frequency = params[1 * n_channels + i]
            amplitude = params[2 * n_channels + i]
            offset = params[3 * n_channels + i]
            
            phases.append(phase)
            frequencies.append(frequency)
            amplitudes.append(amplitude)
            offsets.append(offset)
```

```python
phases_data = np.array(phases[:1000])
np.shape(phases_data)
# phases_data
```

```python
plt.figure(figsize=(12, 6))
data = frequencies[2000:5000]
n_sample = len(frequencies[2000:5000])

x = list(range(n_sample))

# Plot the single line using the index as the x-axis and data values as the y-axis
# plt.plot(phases_data, range(len(phases_data)))
plt.plot(data, label='Phase Line')
# Set the title and labels for x and y axes
# plt.title('Data Line Visualization over Time')
plt.xlabel('Timestamps')
plt.ylabel('Values')

# Show the legend
# plt.legend(loc='upper left', fontsize='small')


# Display the plot
# plt.tight_layout()  # Adjust the layout to fit everything cleanly
plt.show()
```
